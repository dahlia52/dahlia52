### 넘파이 ndarray 개요

- 넘파이 array()함수는 파이썬의 리스트와 같은 다양한 인자를 입력 받아서 ndarray로 변환하는 기능을 수행
- 생성된 ndarray 배열의 shape 변수는 행과 열의 수를 튜플 형태로 가져 배열의 차원까지 알 수 있음

**np.array(), type(array1), array1.shape, array1.dim**

```python
import numpy as np
array1=np.array([1,2,3]) # 1차원
print('array1 type:',type(array1))
print('array1 array 형태:',array1.shape)

array2=np.array([[1,2,3],
               [2,3,4]])
print('array2 type:',type(array2))
print('array2 array 형태:',array2.shape)

array3=np.array([[1,2,3]]) # 2차원
print('array3 type:',type(array3))
print('array3 array 형태:',array3.shape,'\n')

print('array1: {:0}차원, array2: {:1}차원, array3: {:2}차원'.format(array1.ndim,array2.ndim,array3.ndim))
```


### ndarray의 데이터 타입

- ndarray내의 데이터 값은 숫자 값, 문자열 값, 불 값 등 모두 가능
- ndarray내의 데이터 타입은 모두 같은 형태의 데이터 타입만 가능 (리스트와 대비)

```python
# ndarray 내의 데이터 타입은 같은 데이터 타입만 가능
list1=[1,2,3]
print(type(list1))
array1=np.array(list1)
print(type(array1))
print(array1,array1.dtype,'\n')

list2=[1,2,'test']  # 리스트는 다른 형태의 데이터 타입도 가능
array2=np.array(list2)
print(array2,array2.dtype,'\n') # 모두 문자열로 바뀜.

list3=[1,2,3.0]
array3=np.array(list3)
print(array3,array3.dtype) # 모두 float형으로 바뀜.
```

```python
# astype() 메소드로 데이터 타입 변경 -> 메모리 절약
array_int=np.array([1,2,3])
array_float=array_int.astype('float64')
print(array_float,array_float.dtype)

array_int1=array_float.astype('int32')
print(array_int1,array_int1.dtype)

array_float1=np.array([1.1,2.1,3.1])
array_int2=array_float1.astype('int32')
print(array_int2,array_int2.dtype)
```

###ndarray를 편리하게 생성하기 - arange, zeros, ones

- np.arange( )함수는 파이썬 함수 range()와 유사한 기능을 함. 
- 0부터 (함수 인자 값-1)까지의 값을 순차적으로 ndarray의 데이터값으로 변환해줌.
- np.ones( ), np.zeros( )함수는 dtype을 설정하지 않으면 default로 float64형의 데이터로 ndarray를 채움

```python
sequence_array=np.arange(10) # stop값만 부여
print(sequence_array)
print(sequence_array.dtype,sequence_array.shape)

print(np.arange(3,8)) # start값, stop값 부여
```

```python
zero_array=np.zeros((3,2),dtype='int32')
print(zero_array)
print(zero_array.dtype,zero_array.shape,'\n')

one_array=np.ones((3,2)) #dtype을 설정하지 않으면 default로 float64형 데이터
print(one_array)
print(one_array.dtype,one_array.shape)
```

###ndarray의 차원과 크기를 변경하는 reshape( )

- **array1.reshape(행,열)**

- reshape( )에 **-1**값을 인자로 사용하면 원래 ndarray와 호환되는 새로운 shape로 반환해줌 

```python
array1=np.arange(10)
print('array1:\n',array1,'\n')

array2=array1.reshape(2,5)
print('array2:\n',array2,'\n')

array3=array1.reshape(5,2)
print('array3:\n',array3)
```

```python
array1=np.arange(10)
print(array1,'\n')

array2=array1.reshape(-1,5)
print(array2,'\n','array2 shape:',array2.shape,'\n')

array3=array1.reshape(5,-1)
print(array3,'\n','array3 shape:',array3.shape)
```

- reshape(-1,1)은 원본 ndarray가 어떤 형태라도 2차원이고, 1개의 column을 가진 ndarray로 변환됨.

- 여러 개의 넘파이 ndarray는 stack이나 concat으로 결합할 때 각각의 ndarray형태를 통일해 유용하게 사용됨.

- ndarray는 tolist( )메소드를 이용하여 리스트 자료형으로 변환할 수 있음. 때로는 리스트 자료형을 print를 이용해 출력할 때 시각적으로 더 이해하기 쉬울 수 있어서 ndarray를 리스트로 변환해 출력함.

```python
array1=np.arange(8)

array3d=array1.reshape(2,2,2)
print('array3d:\n',array3d.tolist(),'\n')

# 3차원 ndarray를 2차원 ndarray로 변환
array5=array3d.reshape(-1,1)
print('array5:\n',array5.tolist())
print('array5 shape:',array5.shape,'\n')

# 1차원 ndarray를 2차원 ndarray로 변환
array6=array1.reshape(-1,1)
print('array6:\n',array6.tolist())
print('array6 shape:',array6.shape)
```

### 넘파이의 ndarray의 데이터 세트 선택하기 - 인덱싱(Indexing)

1. 특정한 데이터만 추출 : 원하는 위치의 인덱스 값을 지정하면 해당 위치의 데이터가 반환됨.
2. 슬라이싱(Slicing) : 슬라이싱은 연속된 인덱스상의 ndarry를 추출하는 방식

 ':' 기호 사이에 시작 인덱스와 종료 인덱스를 표시하면 시작 인덱스에서 (종료 인덱스 -1) 위치에 있는 데이터의 ndarray를 반환
3. 팬시 인덱싱(Fancy indexing) : 일정한 인덱싱 집합을 리스트 또는 ndarray 형태로 지정해 해당 위치에 잇는 데이터의 ndarray를 반환
4. 불린 인덱싱(Boolean indexing) : 특정 조건에 해당하는지 여부인 True/False 값 인덱싱 집합을 기반으로 True에 해당하는 인덱스 위치에 있는 데이터의 ndarray를 반환

### 단일값 추출

```python
#1부터 9까지의 1차원 ndarray 생성
array1=np.arange(start=1,stop=10)
print('array1:',array1)

#index는 0부터 시작하므로 array1[2]는 3번째 index 위치의 데이터값 의미
value=array1[2]
print('value:',value)
print(type(value))

print('맨 뒤의 값:',array1[-1],'맨 뒤에서 두 번째 값:',array1[-2],'\n')

# 단일 인덱스를 이용하여 ndarray내의 데이터 값을 수정
array1[0]=9
array1[8]=0
print('array1:',array1)
```

```python
# 2차원
array1d=np.arange(1,10)
array2d=array1d.reshape(3,3)
print(array2d)

print('(row=0,col=0) index 가리키는 값 : ', array2d[0,0])
print('(row=0,col=1) index 가리키는 값 : ', array2d[0,1])
print('(row=1,col=0) index 가리키는 값 : ', array2d[1,0])
print('(row=2,col=2) index 가리키는 값 : ', array2d[2,2])
```

### 슬라이싱


```python
# 1차원 슬라이싱
array1=np.arange(start=1,stop=10) # start, stop 생략 가능
array3=array1[0:3]
print(array3,'\n',type(array3))

array4=array1[:3]
print(array4)

array5=array1[3:]
print(array5)

array6=array1[:]
print(array6)
```

```python
# 2차원 슬라이싱
array1d=np.arange(1,10)
array2d=array1d.reshape(3,3)
print('array2d:\n',array2d,'\n')

print('array2d[0:2,0:2]\n',array2d[0:2 , 0:2],'\n')
print('array2d[1:3,0:3]\n',array2d[1:3,0:3],'\n')
print('array2d[1:3,:]\n',array2d[1:3,:],'\n')
print('array2d[:,:]\n',array2d[:,:],'\n')
print('array2d[:2,1:]\n',array2d[:2,1:],'\n')
print('array2d[:2,0]\n',array2d[:2,0],'\n')

print(array2d[0]) # axis 0의 첫번째 row 반환
print(array2d[1]) # axis 0의 두번째 row 반환 
print('array2d[0] shape:',array2d[0].shape,'array2d[1] shape:',array2d[1].shape) # 1차원
```

### 팬시 인덱싱 
- 리스트나 ndarray로 인덱스 집합을 지정하면 해당 위치의 인덱스에 해당하는 ndarray를 반환하는 인덱싱 방식

```python
array1d=np.arange(1,10)
array2d=array1d.reshape(3,3)
print(array2d)

array3=array2d[[0,1],2]
print('array2d[[0,1],2] =>',array3.tolist())

array4=array2d[[0,1],0:2]
print('array2d[[0,1],0:2] =>',array4.tolist())

array5=array2d[[0,1]] # 행은 0,1, row는 전체
print('array2d[[0,1]] =>',array5.tolist())
```

### 불린 인덱싱

- Step 1 : array1d > 5 와 같이 ndarray의 필터링 조건을 [ ] 안에 기재
- Step 2 : False 값은 무시하고 True값에 해당하는 인덱스값만 저장 (유의해야 할 사항은 True값 자체인 1을 저장하는 것이 아니라 True값을 가진 인덱스를 저장한다는 것)
- Step 3 : 저장된 인덱스 데이터 세트로 ndarray 조회

```python
array1d=np.arange(1,10)
# []안에 array1d > 5 Boolean Indexing을 적용
array3=array1d[array1d>5]
print('array1d>5 불린 인덱싱 결과 값:',array3)

boolean_indexing=array1d>5
print(boolean_indexing)
print(array1d[boolean_indexing])
```

### 행렬 정렬

**np.sort( )**: 원 행렬을 유지한 채, 원 행렬의 정렬된 행렬을 반환

**ndarray.sort( )**: 원 행렬 자체를 정렬된 형태로 변환하며 반환 값은 None

내림차순: np.sort( )[::-1]

```python
org_array=np.array([3,1,9,5])
print("원본 행렬:",org_array)

# np.sort()로 정렬
sort_array1=np.sort(org_array)
print('np.sort() 호출 후 반환된 정렬 행렬:',sort_array1) # 원본 행렬을 변경하지 않고 정렬된 형태로 반환/ default 오름차순
print('np.sort() 호출 후 원본 행렬:',org_array,'\n')

# ndarray.sort()로 정렬
sort_array2=org_array.sort()
print('org_array.sort() 호출 후 반환된 정렬 행렬:',sort_array2) # 원본 행렬 자체를 정렬된 값으로 변환/ default 오름차순
print('org_array.sort() 호출 후 원본 행렬:',org_array,'\n')

# 내림차순으로 정렬
sort_array1_desc=np.sort(org_array)[::-1]
print('내림차순으로 정렬:',sort_array1_desc)
```

```python
# 2차원 행렬 정렬
array2d=np.array([[8,12],
                  [7,1]])
sort_array2d_axis0=np.sort(array2d,axis=0)
print('row 방향으로 정렬:\n',sort_array2d_axis0)

sort_array2d_axis1=np.sort(array2d,axis=1)
print('column 방향으로 정렬:\n',sort_array2d_axis1)
```

### 정렬된 행렬의 인덱스를 반환하기

**np.argsort( )**: 정렬 행렬의 원본 행렬 인덱스를 ndarray형으로 반환

-np.argsort()를 이용해 반환된 인덱스를 팬시 인덱스로 적용해 넘파이의 데이터 추출에서 많이 사용됨.

```python
org_array=np.array([3,1,9,5])

sort_indices=np.argsort(org_array)
print(type(sort_indices)) # ndarray형
print('행렬 정렬 시 원본 행렬의 인덱스:',sort_indices)

# 내림차순
sort_indices_desc=np.argsort(org_array)[::-1]
print('행렬 내림차순 정렬 시 원본 행렬의 인덱스:',sort_indices_desc)
```

```python
import numpy as np
name_array=np.array(['John','Mike','Sarah','Kate','Samuel'])
score_array=np.array([78,95,84,98,88])

sort_indices_asc=np.argsort(score_array)
print('성적 오름차순 정렬 시 sort_array의 인덱스',sort_indices_asc)
print('성적 오름차순으로 name_array의 이름 출력:',name_array[sort_indices_asc])
```

### 선형대수 연산 - 행렬 내적과 전치 행렬 구하기

행렬 내적(행렬 곱): **np.dot(A,B)**

전치 행렬: **np.transpose(A)**

```python
# 행렬 내적
A=np.array([[1,2,3],
           [4,5,6]])
B=np.array([[7,8],
            [9,10],
            [11,12]])
dot_product=np.dot(A,B)
print('행렬 내적 결과:\n',dot_product)
```

```python
# 전치 행렬
A=np.array([[1,2],
            [3,4]])
transpose_mat=np.transpose(A)
print('A의 전치행렬:\n',transpose_mat)
```
